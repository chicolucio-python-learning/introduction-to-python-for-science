Your function should take as arguments a `NumPy` array `x` and the order `n`, and should return an array of the designated order `n` spherical Bessel function. Take care to make sure that your functions behave properly at `x = 0`.

Demonstrate the use of your function by writing a Python routine that plots the three Bessel functions for $0 ≤ x ≤ 20$.

```python
def bessel_sph(x, n):
    
    with np.errstate(invalid='ignore', divide='ignore'):  # No warning :-)
        j0 = np.where(x == 0., 1., np.sin(x) / x)
        j1 = np.where(x == 0., 0., (j0 - np.cos(x)) / x)
        j2 = np.where(x == 0., 0., 3. * j1 / x - j0)

    if n == 0:        
        return j0
    elif n == 1:
        return j0, j1
    else:
        return j0, j1, j2


import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.0, 20, 256)
y0, y1, y2 = bessel_sph(x, 2)
plt.plot(x, y0, label='j0(x)')
plt.plot(x, y1, label='j1(x)')
plt.plot(x, y2, label='j2(x)')
plt.legend()
plt.xlabel('x')
plt.axhline(color="gray", zorder=-1)
plt.show()
plt.savefig("besselSph.eps")
```
