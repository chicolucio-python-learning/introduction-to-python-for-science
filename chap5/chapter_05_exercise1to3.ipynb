# Exercício

Plot da função $y = 3x²$ para o intervalo $-1 \leqslant x \leqslant 3$ como uma linha contínua.

```python
### Sempre começar com essas duas linhas para fazer plots
import numpy as np
import matplotlib.pyplot as plt

# criando pontos suficientes para um gráfico smooth
x = np.linspace(-1, 3, 100)  #1 00 pontos parece ser ideal

# escrevendo a função
y = 3 * x**2

# gráfico
plt.figure(1, figsize=(10, 6))
plt.plot(x, y, 'b-',
         label=r'$y=3x^2$')  #usando LaTeX para expressar a legenda.
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.title('plot')
plt.axhline(color="gray", zorder=-1)
plt.axvline(color="gray", zorder=-1)
plt.grid(linestyle=':', linewidth=0.5)
plt.show()

####
#gráfico outro modo
#fig = plt.figure(2, figsize = (10,6))
#ax = fig.add_subplot(1,1,1)
#...
#plt.show()
####
```

# Exercício

Plot da função $y = \dfrac{\cos{x}}{1+\frac{1}{5}x²}$ para o intervalo $-15 \leqslant x \leqslant 15$ como uma linha contínua.

```python
### Sempre começar com essas duas linhas para fazer plots
import numpy as np
import matplotlib.pyplot as plt

#criando pontos suficientes para um gráfico smooth
x = np.linspace(-15, 15, 200)

#escrevendo a função
y = np.cos(x) / (1 + 1 / 5 * x**2)

#gráfico
plt.figure(1, figsize=(10, 6))
plt.plot(x, y, 'b-', label="y")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.title('plot')
plt.axhline(color="gray", zorder=-1)
plt.axvline(color="gray", zorder=-1)
plt.grid(linestyle=':', linewidth=0.5)
plt.show()
```

# Exercício

Plot das funções $\sin{x}$ e $\cos{x}$ versus $x$.

```python
### Sempre começar com essas duas linhas para fazer plots
import numpy as np
import matplotlib.pyplot as plt

#criando pontos suficientes para um gráfico smooth
x = np.linspace(-np.pi, np.pi, 100)

#escrevendo a função
ysin = np.sin(x)
ycos = np.cos(x)

#gráfico
plt.figure(1, figsize=(10, 6))
plt.plot(x, ysin, 'g-', label="sin(x)")
plt.plot(x, ycos, 'k-', label="cos(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-np.pi, np.pi)
plt.legend(loc='upper right')
plt.title('plot')
plt.axhline(color="gray", zorder=-1)
plt.axvline(color="gray", zorder=-1)
plt.grid(linestyle=':', linewidth=0.5)
plt.show()
```
