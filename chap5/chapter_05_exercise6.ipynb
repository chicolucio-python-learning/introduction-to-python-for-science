```python
### Sempre começar com essas duas linhas para fazer plots
import numpy as np
import matplotlib.pyplot as plt

#data
d = np.array([0.38, 0.64, 0.91, 1.26, 1.41, 1.66, 1.90, 2.18])
f = np.array([1.4, 1.65, 3.0, 3.95, 4.3, 5.20, 6.85, 7.4])
df = np.array([0.4, 0.5, 0.4, 0.5, 0.6, 0.5, 0.5, 0.4])

#fit
fit = np.polyfit(d, f, 1)
fit_fn = np.poly1d(fit)

# create plot
plt.figure(1, figsize=(10, 6))
(_, caps, _) = plt.errorbar(d,
                            f,
                            fmt='yd',
                            marker='d',
                            markerfacecolor='yellow',
                            markersize=12,
                            markeredgecolor='red',
                            yerr=df,
                            ecolor='red',
                            capsize=5)

plt.plot(d, fit_fn(d), 'b-')

plt.xlim(0, 2.5)
plt.ylim(0, 9)

plt.xlabel('distance (cm)')
plt.ylabel('force (N)')
#plt.legend(loc='upper right')

for cap in caps:
    cap.set_markeredgewidth(1)

# display plot on screen
plt.show()
```

```python
### Mesmo exercício usando scipy com linregress

### Sempre começar com essas duas linhas para fazer plots
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#data
d = np.array([0.38, 0.64, 0.91, 1.26, 1.41, 1.66, 1.90, 2.18])
f = np.array([1.4, 1.65, 3.0, 3.95, 4.3, 5.20, 6.85, 7.4])
df = np.array([0.4, 0.5, 0.4, 0.5, 0.6, 0.5, 0.5, 0.4])

#fits mínimo e máximo
d_min = np.array([d[0], d[7]])
d_max = d_min
f_min = np.array([f[0] - df[0], f[7] + df[7]])
f_max = np.array([f[0] + df[0], f[7] - df[7]])

#fit
slope, intercept, r_value, p_value, std_err = stats.linregress(d, f)

# create plot
plt.figure(1, figsize=(10, 6))
(_, caps, _) = plt.errorbar(d,
                            f,
                            fmt='yd',
                            marker='d',
                            markerfacecolor='yellow',
                            markersize=12,
                            markeredgecolor='red',
                            yerr=df,
                            ecolor='red',
                            capsize=5)

plt.plot(d, intercept + slope * d, 'b-')

plt.plot(d_min, f_min, 'c--')

plt.plot(d_max, f_max, 'g--')

plt.xlim(0, 2.5)
plt.ylim(0, 9)

plt.xlabel('distance (cm)')
plt.ylabel('force (N)')
#plt.legend(loc='upper right')

for cap in caps:
    cap.set_markeredgewidth(1)

# display plot on screen
plt.show()
```
